```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../cleaned_data/aaa_ev_prices_comparison.csv')
```

```python
display(data)
```

```python
new_data = data.dropna()

display(new_data)
```

```python
new_data['Australia'] = new_data['Australia'].apply(lambda x: x.split(' ')[0].replace('A$', '').replace(',', '')).apply(pd.to_numeric)
new_data['Norway'] = new_data['Norway'].apply(lambda x: x.split(' ')[0].replace('A$', '').replace(',', '')).apply(pd.to_numeric)
new_data['New Zealand'] = new_data['New Zealand'].apply(lambda x: x.split(' ')[0].replace('A$', '').replace(',', '')).apply(pd.to_numeric)
new_data['Germany'] = new_data['Germany'].apply(lambda x: x.split(' ')[0].replace('A$', '').replace(',', '')).apply(pd.to_numeric)
new_data['China'] = new_data['China'].apply(lambda x: x.split(' ')[0].replace('A$', '').replace(',', '')).apply(pd.to_numeric)
new_data['Japan'] = new_data['Japan'].apply(lambda x: x.split(' ')[0].replace('A$', '').replace(',', '')).apply(pd.to_numeric)
new_data['USA'] = new_data['USA'].apply(lambda x: x.split(' ')[0].replace('A$', '').replace(',', '')).apply(pd.to_numeric)
new_data['Great Britain'] = new_data['Great Britain'].apply(lambda x: x.split(' ')[0].replace('A$', '').replace(',', '')).apply(pd.to_numeric)

display(new_data)
```

```python
df = pd.DataFrame({
    'Model': ['Porsche Taycan', 'Tesla Model 3', 'Tesla Model Y', 'Toyota RAV4 Hybrid', 'Volvo XC40 BEV'],
    'Australia': [158100, 65500, 100235, 36900, 72990],
    'Norway': [120084, 61979, 84523, 76875, 69135],
    'New Zealand': [165189, 67499, 98140, 42617, 75610],
    'Germany': [137673, 82488, 93390, 58387, 75768],
    'China': [195906, 63481, 86150, 49260, 60722],
    'Japan': [130260, 63366, 88536, 37590, 61517],
    'USA': [130348, 74063, 92977, 46614, 81487],
    'Great Britain': [135392, 85180, 120149, 61669, 80518]
})
```

```python
# Melt the DataFrame to make it suitable for seaborn
df_melted = df.melt(id_vars='Model', var_name='Country', value_name='Price')

# Create the barplot
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='Price', hue='Country', data=df_melted)
plt.title('Price Comparison of Car Models by Country')
plt.xlabel('EV Model')
plt.ylabel('Price (Australian Dollar)')
plt.show()
```
