```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../cleaned_data/abs_automotive_fuel_and_CPI_Index.csv')
```

```python
display(data)
```

```python
data = data.dropna()
display(data)
```

```python
data["Automotive Fuel"] = data["Automotive Fuel"].apply(lambda x: x.replace(',', ''))
data["CPI"] = data["CPI"].apply(lambda x: x.replace(',', ''))
```

```python
data["Automotive Fuel"] = data["Automotive Fuel"].apply(pd.to_numeric)
data["CPI"] = data["CPI"].apply(pd.to_numeric)


data.info()
```

```python
# Write the cleaned dataframe to CSV without including the index column
data.to_csv("../cleaned_data/cleaned_automotive_fuel_and_CPI_Index.csv", index=False)
```

```python
plt.figure(figsize=(14, 8))

# Draw line plot
sns.lineplot(x='Year', y='value', hue='variable',
             data=pd.melt(data, ['Year']))

plt.title('Automotive fuel and CPI Index Number, 1972 = 100.0')
plt.xticks(rotation=50)
plt.show()
```
