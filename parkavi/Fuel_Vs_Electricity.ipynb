```python
#Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Read in Annual Electricity Prices CSV

Electricity_path = "../Resources/Annual_Electricity_Prices.csv"
Electricity_df = pd.read_csv(Electricity_path)
Electricity_df
```

```python
Electricity_df.info()
```

```python
# Read in Annual Diesel Prices CSV

Diesel_path = "../Resources/AIP_Annual_TGP_Diesel.csv"
Diesel_df = pd.read_csv(Diesel_path)
Diesel_df
```

```python
Diesel_df.info()
```

```python
# Read in Annual Petrol Prices CSV

Petrol_path = "../Resources/AIP_Annual_TGP_Petrol.csv"
Petrol_df = pd.read_csv(Petrol_path)
Petrol_df
```

```python
Petrol_df.info()
```

```python
# Merge Petrol_df and Diesel_df on 'Financial Year'
merged_df = pd.merge(Petrol_df, Diesel_df, on='Financial Year', suffixes=('_Petrol', '_Diesel'))

# Merge the result with Electricity_df on 'Financial Year'
final_merged_df = pd.merge(merged_df, Electricity_df, on='Financial Year')

# Display the merged DataFrame
print(final_merged_df.head())
```

```python
final_merged_df.info()
```

```python
Fuel_Electricity_df=final_merged_df[["Financial Year", "National Average(Petrol)", "National Average(Diesel)", "National Average($/MWh)"]]
Fuel_Electricity_df.info()
```

```python
# Extract the relevant columns
financial_years = Fuel_Electricity_df['Financial Year']
petrol_avg = Fuel_Electricity_df['National Average(Petrol)']
diesel_avg = Fuel_Electricity_df['National Average(Diesel)']
electricity_avg = Fuel_Electricity_df['National Average($/MWh)']

# Define the number of bars per group
bar_width = 0.25

# Set position of bars on X axis
r1 = np.arange(len(financial_years))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Create the figure and the bar chart
plt.figure(figsize=(15, 8))

# Plotting the bars for each category
plt.bar(r1, petrol_avg, color='blue', width=bar_width, edgecolor='grey', label='Petrol')
plt.bar(r2, diesel_avg, color='orange', width=bar_width, edgecolor='grey', label='Diesel')
plt.bar(r3, electricity_avg, color='green', width=bar_width, edgecolor='grey', label='Electricity')

# Add xticks on the middle of the group bars
plt.xlabel('Financial Year', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(financial_years))], financial_years, rotation=45)

# Add labels, title, and legend
plt.ylabel('Price in AUD')
plt.title('National Average Prices of Petrol, Diesel, and Electricity by Financial Year')
plt.legend()

# Show the plot
plt.tight_layout()  # Adjust layout to make room for rotated x labels
plt.show()

```
