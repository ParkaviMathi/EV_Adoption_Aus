```python
import requests
# Request for the api

# https://api.openchargemap.io/v3/poi?key=2f4c4594-973b-40b4-a9e6-fc7c2404c9fd&countrycode=AU&maxresults=1000

r = requests.get('https://api.openchargemap.io/v3/poi?key=2f4c4594-973b-40b4-a9e6-fc7c2404c9fd&countrycode=AU&maxresults=2000')

data = r.json()
```

```python
save_json = "openchargemap_australia_full_2000.json"
with open (save_json, 'w') as outfile:
    json.dump(data, outfile, indent=4)
    print(f"Data saved successfully to {save_json}")
```

```python
len(data)
```

```python
lats_found = []
longs_found = []

for i in range(len(data)):
    info = data[i]['AddressInfo']

    lat = float(info['Latitude'])
    long = float(info['Longitude'])
    print(f"{lat} || {long}")

    
    lats_found.append(lat)
    longs_found.append(long)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Latitude": lats_found,
    "Longitude": longs_found
})

df.head()
```

```python
save_path = "cleaned_data/ev_charge_stations_australia.csv"
df.to_csv(save_path)
print(f"Successfully saved to {save_path}")
```
