```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://ev-database.org/cheatsheet/range-electric-car'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
"""
<div class="core-content">
<section>
    <table>
        <tbody>
            <tr>

"""

None
```

```python
soup_query_1 = soup.find_all('table')[0]
soup_query_1
```

```python
soup_query = soup_query_1.find_all('tr')
soup_query
```

```python

for entry in soup_query[1:]: # Skip first row
    # print(entry)
    car_name = entry.find('a').string
    car_range = entry.find('span').string

    print(f"Car: {car_name}   ||   Range: {car_range}")
    print(type(car_name))
    break
```

```python


vehicle_ls = []
range_ls = []

for entry in soup_query[1:]: # Skip first row
    # print(entry)
    car_name = entry.find('a').string
    car_range = entry.find('span').string

    print(f"Car: {car_name}   ||   Range: {car_range}")
    vehicle_ls.append(car_name)
    range_ls.append(int(car_range))
    
```

```python
vehicle_range_dict = {
    "Vehicle":vehicle_ls,
    "Range (km)": range_ls
}

```

```python
import pandas as pd

df = pd.DataFrame.from_dict(vehicle_range_dict)
df.head(20)
```

```python
save_csv_path = "cleaned_data/electric_vehicle_ranges_ev_database.csv"
df.to_csv(save_csv_path)
```
