```python
# Read excel file: sheet_name='all_alpha_24'

in_path = 'Resources/SmartWay Vehicle List for MY 2024.xlsx'
vehicle_mpg_df = pd.read_excel(in_path, sheet_name='all_alpha_24')
vehicle_mpg_df.head()
```

```python
# Remove columns:
remove_cols = ['Displ', 'Cyl', 'Trans', 'Drive', 'Cert Region', 'Stnd',
               'Stnd Description', 'Underhood ID', 'Air Pollution Score',
               'Greenhouse Gas Score', 'SmartWay',
               'City MPG', 'Hwy MPG']
```

```python
vehicle_mpg_df = vehicle_mpg_df.drop(columns=remove_cols)
vehicle_mpg_df.head()
```

```python
convert_mpg_to_l_per_100_km = 235.214583 # 1 mile per gallon is equal to 235.214583 liters per 100 km

l_per_100_km_ls = []
for row in vehicle_mpg_df.iterrows():
    data = row[1]

    try:
        l_per_100_km = convert_mpg_to_l_per_100_km / float(data['Cmb MPG']) 
    except:
        l_per_100_km = None
    l_per_100_km_ls.append(l_per_100_km)
```

```python
vehicle_efficiency_df = vehicle_mpg_df.copy()
vehicle_efficiency_df.insert(3, "L/100km", l_per_100_km_ls, allow_duplicates=True)
```

```python
vehicle_efficiency_df.head()
```

```python
# Finally drop 'Cmb MPG'
vehicle_efficiency_df = vehicle_efficiency_df.drop('Cmb MPG', axis=1)
vehicle_efficiency_df.head()
```

```python
# Save the dataframe to csv

save_path = 'cleaned_data/vehicle_fuel_efficiency.csv'
vehicle_efficiency_df.to_csv(save_path)
print(f"Saved file to {save_path}")
```
